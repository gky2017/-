```python
# sin함수만 그려보기
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
# cos 함수도 추가해보기
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, linestyle = '--', label = 'cos')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin & Cos Graph')
plt.legend()
plt.show()
```
